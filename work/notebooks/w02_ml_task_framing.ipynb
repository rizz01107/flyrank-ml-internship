## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

In an earlier notebook, a depth-2 decision tree beat a hand-written rule at Precision@50, showing the model finds combinations of signals (staleness + visibility + position) that a single fixed threshold misses. ML lets the priority list adapt across many correlated signals instead of one manual cutoff.
